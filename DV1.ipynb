```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
from matplotlib import style
style.use('ggplot') #background= grey
```

```python
#there are some inbuilt dataset in seaborn libraby like,
snr.get_dataset_names()
```

```python
#let's use 'iris'
iris=snr.load_dataset('iris')
iris
```

```python
#dropping 'species' column as it's categorical in nature, not numerical.
data=iris.drop('species',axis=1) #axis=1 used to drop entire 'species column and axis=0 to drop entire row
data
```

```python
data.describe()
```

#MATPLOTLIB

```python
#LINEGRAPH/LINEPLOT
linegraph=data.plot(figsize=(15,7),title='iris') #give any variable to upload linegraph, here= 'linegraph',figsize=size of graph, title=name of graph
linegraph.set_xlabel('x axis') #x axis name
linegraph.set_ylabel('y label') #y axis name
```

```python
#HISTOGRAM/DISTPLOT
hist=data.plot(kind='hist',figsize=(15,7),bins=150,title='iris')
```

```python
hist=data.diff().plot(kind='hist',figsize=(15,7),bins=150,title='iris') #diff()
```

```python
data['sepal_length'].diff().plot(kind='hist',figsize=(10,5),bins=150)
```

```python
data['sepal_width'].diff().plot(kind='hist',figsize=(10,5),bins=100)
```

```python
data['petal_length'].diff().plot(kind='hist',figsize=(10,5),bins=100)
```

```python
data['petal_width'].diff().plot(kind='hist',figsize=(10,5),bins=100)
```

```python
#BARGRAPH/BARPLOT
bar=data.iloc[0:21].plot(kind='bar',figsize=(20,7)) #for 1st 20 rows
bar.set_xlabel('x-axis')
bar.set_ylabel('y-axis')
```

```python
data.iloc[0:10,0:2].plot(kind='barh',figsize=(20,7)) #for 1st 20 rows,#barh=horizontal
```

```python
#SCATTERPLOT- used to compare 2-3 parameters at a time
data.plot.scatter(x='sepal_length',y='sepal_width',figsize=(15,7)) #blue is by default colour
```

```python
data.plot.scatter(x='sepal_length',y='sepal_width',c='r',figsize=(15,7),title='iris') #c=colour
```

```python
data.plot.scatter(x='sepal_length',y='sepal_width',c='petal_length',figsize=(15,7),title='iris') #comparing 3 parameters
```

```python
#PIE CHART
data.iloc[0].plot.pie()
```

```python
data.iloc[0].plot.pie(autopct='%.1f',figsize=(7,7)) #autopct(auto percentage)=upto 1 decimal points
```

```python
data.iloc[0:10].plot.pie(subplots=True,figsize=(15,10)) #subplots=True- for multiple pie charts
```

```python
#savefig- saves file 
plt.pie([40,30,20])
plt.savefig('pie') #saved above file in 'png' format by default,we can save it to .pdf, .csv etc
plt.show()
```

#SEABORN

```python
#now let's use 'tips' dataset for further visualizations
tips=snr.load_dataset('tips')
tips
```

```python
tips.describe()
```

```python
#LINEPLOT/LINEGRAPH
snr.lineplot(x='total_bill',y='tip',data=tips)
```

```python
plt.figure(figsize=(20,8))
snr.lineplot(x='total_bill',y='tip',data=tips)
```

```python
plt.figure(figsize=(20,8))
snr.lineplot(x='size',y='total_bill',data=tips,hue='sex',)
plt.show()
```

```python
#HISTOGRAM/DSTPLOT
snr.distplot(tips['total_bill'])
```

```python
plt.figure(figsize=(20,8))
snr.distplot(tips['tip'],bins=100)
```

```python
plt.figure(figsize=(20,8))
snr.distplot(tips['tip'],bins=100,hist=False) #hist=False, hides histogram
```

```python
plt.figure(figsize=(20,8))
snr.distplot(tips['tip'],bins=100,kde=False) #kde=False, hides line
```

```python
#BARGRAPH/BARPLOT
snr.barplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#SCATTERPLOT
plt.figure(figsize=(20,8))
snr.scatterplot(y='total_bill',x='tip',data=tips,hue='sex',style='time')
```

```python
plt.figure(figsize=(20,8))
snr.scatterplot(y='total_bill',x='tip',data=tips,hue='sex',style='time',size='size',sizes=(100,500)) #sizes=to expand size with min value to max value 
```

```python
#HEATMAP- graphical representation of 2D dataset.each datavalue represents in a matrix with special colour 
#so, let's create 2D  dataset
number=np.linspace(1,100,50).reshape(10,5)
number
```

```python
snr.heatmap(number)
```

```python
plt.figure(figsize=(15,8))
snr.heatmap(data)
```

```python
#RELATIONALPLOT
snr.relplot(x='total_bill',y='tip',data=tips)
```

```python
snr.relplot(x='total_bill',y='tip',data=tips,col='sex',hue='time',size='size',style='day',sizes=(50,150))
```

```python
#CATPLOT
snr.catplot(x='total_bill',y='day',data=tips)
```

```python
#BOXPLOT
snr.catplot(x='day',y='tip',data=tips,kind='box',hue='smoker')
```

```python
#PAIRPLOT
snr.pairplot(tips, x_vars=['tip','day','size'],y_vars=['tip'],size=5)
```
